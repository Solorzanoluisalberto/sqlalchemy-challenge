```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy.engine import reflection
from sqlalchemy import create_engine, func, inspect
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
# ==> Assign the measurement class to a variable called `measure`
# ==> Assign the station class to a variable called `stn`
measure = Base.classes.measurement
stn = Base.classes.station
engine.execute('SELECT * FROM measurement LIMIT 5').fetchall()
```

```python
inspector = inspect(engine)
columns = inspector.get_columns('measurement')
for c in columns:
    print(c['name'], c["type"])
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
first_row = session.query(measure).first()
first_row.__dict__
```

```python
first_row = session.query(stn).first()
first_row.__dict__
```

```python
# Find the most recent date in the data set.
recent_date = session.query(measure.date).order_by(measure.date.desc()).first()
print(f'The most recent date is: {recent_date[0]}')
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 

# Calculate the date one year from the last date in data set.
dates = dt.date(2017, 8, 23) - dt.timedelta(days=365)
dates # datetime.date(2016, 8, 23)
# Perform a query to retrieve the data and precipitation scores
results = session.query(measure.date, func.avg(measure.prcp)).\
            filter(measure.date.between (dates, dt.date(2017, 8, 23))).group_by('date')
# results
for vaue in results:
    print(vaue)
```

```python
# # Save the query results as a Pandas DataFrame and set the index to the date column
my_df = pd.DataFrame(results, columns=['date', 'precipitation'])
# set the index to the date column.
my_df.set_index('date', inplace=True)
# df.columns = resoverall.keys()
# Sort the dataframe by date
my_df.sort_values(by='date')
my_df
```

```python
# * Plot the results using the DataFrame `plot` method.
my_df.plot(figsize= (10,6), rot = 90);
plt.ylabel("Precipitation (in)", size=14);
plt.title("Precipitation in 12 Months from 2016-08-23 to 2017-08-23", size=20)
plt.savefig("output/plot_date_vs_prcp.png")
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
my_df.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
total_number_stations  = session.query(stn.station.distinct()).filter(stn.station.isnot(None)).count()
total_number_stations
print(f'The total number of stations in the data set is: {total_number_stations}')
# for station in total_number_stations:
#     print(station)
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
# active_stations = session.query(measure.station, func.count(measure.station)).order_by(func.count(measure.station).desc()).filter(measure.station.isnot(None), measure.date.between(dates, dt.date(2017, 8, 23))).group_by(measure.station)

active_stations = session.query(measure.station, stn.name, func.count(measure.station))\
                .join(stn, stn.station == measure.station).order_by(func.count(measure.station).desc())\
                .filter(measure.station.isnot(None))\
                .group_by(measure.station, stn.name)
# most_active_stations;
for station in active_stations:
    print(station)

print(f'The most active stations is:{active_stations[0][0]} {active_stations[0][1]}')
print(f'with {active_stations[0][2]} rows')
str(active_stations)
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
AVG_most_active_station = session.query(func.min(measure.tobs), func.max(measure.tobs), func.avg(measure.tobs))\
                          .filter(measure.station == active_stations[0][0]).all()
AVG_most_active_station

print(f'Lowest temperature: {AVG_most_active_station[0][0]}')
print(f'Highest temperature: {AVG_most_active_station[0][1]}')
print(f'Average temperature: {round(AVG_most_active_station[0][2],2)}')
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
temp_obser_12 = session.query(measure.date, measure.tobs)\
                .filter(measure.station == active_stations[0][0], measure.date.between(dates, dt.date(2017, 8, 23)))\
                .group_by(measure.date)\
                .order_by(measure.date)

for station in temp_obser_12:
    print(station)            
    
str(temp_obser_12)
```

```python
temp_obser_df = pd.DataFrame(temp_obser_12, columns=['Date', 'Temp'])
temp_obser_df.set_index('Date', inplace=True)
plt.xlabel('Temperature', fontsize='12')
plt.ylabel('Frec. temperature', fontsize='12')
plt.title(f'Temperature observation for: {active_stations[0][0]}', fontsize='12')
plt.hist(temp_obser_df, bins=12, alpha=.9, color='#00177f', label='temperature', edgecolor='w')
plt.grid(axis='x', alpha=.4)
plt.legend();
```

```python
from scipy.stats import norm
domain = np.linspace(temp_obser_df['Temp'].min(), temp_obser_df['Temp'].max())
mean_temp = temp_obser_df['Temp'].mean()
std_temp = temp_obser_df['Temp'].std()
plt.plot(domain, norm.pdf(domain, mean_temp, std_temp),
         label= '$\mathcal{N}$ ' + f'$( \mu \\approx {round(mean_temp)} , \sigma \\ approx {round(std_temp)} )$', color='black')
plt.hist(temp_obser_df, edgecolor='w', alpha=.9, density=True, color='#00177f')
plt.xlabel('Temperature', fontsize='12')
plt.ylabel('Frec. temperature', fontsize='12')
plt.title(f'Normal fit Temperature observation for: {active_stations[0][0]}', fontsize='12')
plt.grid(axis='x', alpha=.4)
plt.legend()
plt.show();
print(f'The average temperature of station {active_stations[0][0]} is: {round(mean_temp,2)}')
```
